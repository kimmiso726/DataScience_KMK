카운터로 *세기*

딕셔너리 스타일의 컬렉션에서 다른 컬렉션 안의 아이템 개수 셀 때 사용

```python
from collections import Counter

phrase = "a man a plan a canal panama"
cntr = Counter(phrase.split())

cntr

# phrase.split()으로 잘랐을 때 리스트로 반환 -> 리스트는 순서를 가지고 있음
# 리스트의 순서대로 Counter(리스트) 했을 때 순서가 된다.
```

```python
from collections import Counter

phrase = "a man a plan a canal panama"
cntr = Counter(phrase.split())

cntr.most_common(2)  # 가장 빈번하게 등장한 n개의 아이템과 빈도 추출
```

출력한 리스트를 딕셔너리로 변환하면 가독성이 좋아짐

```python
cntDict = dict(cntr.most_common())
cntDict
```

```python
cntDict['a']
```

```python
# search 문자열 일부분이 정규식과 매칭되는지 확인
# 매칭되는게 앞부분에 없다면 이 함수를 사용한다.
# [a-z] : 소문자부터 매치

re.search(r"[a-z]+", "01100110 has at test one 010 letter 0010010", re.I).group()
```

```python
# search 문자열 일부분이 정규식과 매칭이 되는지 확인
# (매칭되는게 앞부분에 없다면 이 함수 사용)
# [a-z] : 소문자들만 일치
import re
re.search(r"[a-z]+", "01100110 has at test one 010 letter 0010010").group()
```

```python
# search 문자열 일부분이 정규식과 매칭되는지 확인
# 매칭되는게 앞부분에 없다면 이 함수 사용
# [A~Z] : 대문자들만 일치
import re
re.search(r"[A-Z]+", "01100110 has at test one 010 letter 0010010", re.I).group()
```

```python
# finall : 정규식에 부합하는 모든 문자열 찾기(대소문자 무시)
# --> 모든 소문자 들어 있는 토큰 찾기
re.findall(r"[a-z]+", "01100110 has at test one 010 letter 0010010")
```

```python
# findall : 정규식에 부합하는 모든 문자열 찾기 -> 모든 소문자 들어 있는 토큰 찾기
re.findall(r"\d+", "01100110 has at test one 010 letter 0010010")
```

```python
# findall : 정규식에 부합하는 모든 문자열 찾기 -> 모든 소문자 들어 있는 토큰 찾기
re.findall(r"\D+", "01100110 has at test one 010 letter 0010010")
```

```python
re.findall(r"[^a-z]+", "01100110 has at test one 010 letter 0010010")
```

```python
# sub: string으로 매칭되는 부분 문자열을 repl로 치환
# 패턴, 치환 전, 치환 후 값을 줘야 함
re.sub(r"[a-z]+", '[...]', "01100110 has at test one 010 letter 0010010")
```

```python
# sub: string으로 매칭되는 부분 문자열을 repl로 치환
# 패턴, 치환 전, 치환 후 값을 줘야 함
re.sub(r"[a-z]+", '[...]', "01100110 has at test one 010 letter 0010010", flags=re.I)
# flags=re.I는 대소문자 구분 안하도록 하는 것
# re.sub(패턴, 리플, 스트링, 카운트=0, 플래그=0) <- 자리 순서에 맞춰서 넣어야 하기 때문에 여기서는 flags=이라는 지정을 따로 해줘야 했음
```

```python
# 문제1) 010-abcd-0100 -> abcd 값을 ****으로 치환 (re.sub 통해서 만들기)

re.sub(r"[a-z]+", '****', '010-AbcD-0100', flags=re.I)
```

```python
# 문제2) 010-abcd-0100 -> 숫자를 *로 치환
re.sub(r"\d+", "****", "010-abcd-0100")
```

```python
# 문제2) 010-abcd-0100 -> 숫자를 *로 치환
re.sub(r"\D+", "****", "010-abcd-0100")
# 하이픈까지 사라진 것 확인
```
